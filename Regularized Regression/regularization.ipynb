## Calculating Cost Function for 3 different possible fit

Fit 1 has slope 0.5, Fit 2 has slope 1.0, Fit 3 has slope 1.5

Calculating the cost function for each slope, we get the lowest for Fit 2. 

```python
import numpy as np

X = np.array([[1], [2], [3]])
y = np.array([[1], [2.5], [3.5]])

get_theta = lambda theta: np.array([[0, theta]])

thetas = list(map(get_theta, [0.5, 1.0, 1.5]))

X = np.hstack([np.ones([3, 1]), X])

def cost(X, y, theta):
    inner = np.power(((X @ theta.T) - y), 2)
    return np.sum(inner) / (2 * len(X))

for i in range(len(thetas)):
    print(cost(X, y, thetas[i]))
```

```python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import Ridge
```

```python
## Create datasets
A, b, coefficients = make_regression(
    n_samples=50,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=5,
    coef=True,
    random_state=1
)
```

```python
alpha = 1
n, m = A.shape

I = np.identity(m)
w = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A) + alpha * I), A.T), b)
```

```python
plt.scatter(A, b)
plt.plot(A, w*A, c='red')
```

```python
## ridge with alpha 0.5
rr = Ridge(alpha=0.5)
rr.fit(A, b)
w = rr.coef_

plt.scatter(A, b)
plt.plot(A, w*A, c='red')
```

```python
## ridge with alpha 10
rr = Ridge(alpha=10)
rr.fit(A, b)
w = rr.coef_[0]
plt.scatter(A, b)
plt.plot(A, w*A, c='red')
```

```python
## ridge with alpha 100
rr = Ridge(alpha=100)
rr.fit(A, b)
w = rr.coef_[0]
plt.scatter(A, b)
plt.plot(A, w*A, c='red')
```
